13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó 

```python
mean_age=df['AGE'].mean()
df['AGE']=df['AGE'].apply(lambda x: mean_age if x > 40 else x)
df
```

14. Biết công thức tính chỉ số BMI như sau: 
BMI = Weight(kg)/Height2 (m2)  
Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BIM mỗi người

```python
df['cs_BMI']=df['Weight'] / (df['Height']*df['Height'])
df
```

15. Theo thang đo kết quả tính BMI của châu Á: 
Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình  
cân, từ 25 trở lên là béo phì. 
thường, 23-24,9 là thừa 
Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ 
số BMI trên.

```python
df['Phan_Loai']=df['cs_BMI'].apply(lambda x: 0 if x <18.5 else( 1 if 18.5< x <22.9 else( 2 if 23< x<24.9 else 3) ) )
df
```

16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1. 

```python
BMI_max=df['cs_BMI'].max()
BMI_min=df['cs_BMI'].min()
df['cs_BMI']=(df['cs_BMI']-BMI_min)/(BMI_max-BMI_min)
df
```

17. Cho biết: 
+ ‘Mr‘tương ứng với ‘male’, 
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’.  
Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong 
cột <Name> còn các từ đó.

```python
df['GT']=df['Name'].apply(lambda x: 'Male' if "Mr." in x else 'Female')
df['Name']=df['Name'].str.replace('Mr. |Mrs. |Miss. ', ' ',regex=True).str.strip()
df
```

18. Vẽ biểu đồ tròn cho <GT>

19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>.

```python
female_df=df[df['GT']=='Female']
counts=female_df['Phan_Loai'].value_counts()
counts.plot(kind='bar')
plt.xlabel('Phan loai')
plt.ylabel('So luong')
plt.show()
```
